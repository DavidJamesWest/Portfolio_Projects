## Visualisations

Investigate the information and produce some related visualisations

```python
sns.barplot(x=file['Team'],y=file['Tries'])
plt.axhline(file['Tries'].mean(), linestyle = '--', color = 'black', label = 'Mean')
plt.title('Number of Tries for Each Country')
plt.legend()
plt.show()
```

The above visualisation demonstrates the disparity of tries scored, with France, Ireland, and Scotland clearly outscoring England, Wales, and Italy.

```python
sns.scatterplot(x=file['Metres'],y=file['Metres Kicked'], hue=file['Team'], s = 100)
plt.axvline(x=file['Metres'].mean(), linestyle = '--', label = 'Mean')
plt.axhline(y=file['Metres Kicked'].mean(),linestyle = '--')
plt.title('Metres Carried vs Metres Kicked')
plt.xlabel('Metres Carried')
plt.ylabel('Metres Kicked')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

The above visualisation demonstrates the difference in playstyle as relates to metres gained by carries and kicks. 
Ireland are above the mean in both metres carried and kicked, suggesting a competancy and willing to perform both actions.
England and France have very similar outputs for both metrics, implying the ratio of gaining metres is similar between the two countries.
Scotland and Wales have similar output for kicking metres but have very different levels for metres carried.
Italy has a much lower metres kicked output, suggesting they are either ineffective or discouraged from kicking the ball.
The above visualisation could be improved if possession statistics were available as they would allow for a level of normalization between the different teams.

```python
file[['Team','Metres','Metres Kicked']].sort_values(['Metres'], ascending = False).plot(x = 'Team', kind = 'bar', stacked = True)
plt.ylabel('Metres')
plt.xticks(rotation = 45)
plt.title('Metres Carried and Kicked for each Country')
plt.legend(bbox_to_anchor = (1,1))
plt.show()
```

The above visualisation demonstrates the total metres made by each country broken down by carries and kicks. Ireland, France, and England all had similar values for total metres made but Ireland carried for more of those metres. Whilst Italy has the lowest combined total, being approximately 2000 metres behind Ireland.

```python
file[['Team','Metres','Metres Kicked']].plot(x = 'Team', kind = 'bar')
plt.ylabel('Metres')
plt.xticks(rotation = 45)
plt.title('Metres Carried and Kicked for each Country')
plt.legend(bbox_to_anchor = (1,1))
plt.show()
```

The above visualisation shows the comparison of metres carried vs metres kicked for each country. England and France show the greatest discrepancy between carrying and kicking, with kicking being much more dominant. Italy are the only country to carry for more metres thand they kicked. Scotland are the most even country between kicking and carrying.

```python
file_corr = file[['Tries','Metres','Metres Kicked', 'Carries', 'Tackles Made']].copy()
file_corr.corr()
```

```python
sns.heatmap(file_corr.corr())
plt.title('Heatmap for Different Metrics')
plt.show()
```

The above heatmap visualisation provides the correlation between different metrics.
Metres Carried and Metres Kicked are positively correlated. This suggests the team that carries the ball more, will also kick the ball for more metres.
Carries and Tackles Made are negatively correlated. This suggests the more carries you make the fewer tackles you have to make.

```python
file_players['Mins Played'].plot(kind='hist', bins = 20, edgecolor = 'black')
plt.title('Histogram of player minutes in 2023 Six Nations')
plt.xlabel('Minutes')
plt.ylabel('Count')
plt.show()
```

The above visualisation demonstrates a histogram for the minutes players played with a maximum value of 400 mins, with bins of width 20. 
The largest bin is the 380-400 min bin, which correlates with the desire to have the best players playing as many minutes as possible. However, the bin only contained 21 players which will be in part due to rotation and injuries, but does demonstrate the lack of must play players in each team.
The next largest bin is the 0-20 min bin, which suggests a number of players were given small tastes of playing international rugby.

```python
sns.stripplot(file_players, x='Team', y='Mins Played', hue = 'Team')
plt.legend(bbox_to_anchor = (1,1))
plt.title('Jitter plot of Distribution of Mins Played for each Country')
plt.show()
```

The above visualisation is created to help show the distribution of players minutes for each country. The low number of players makes it harder to identify clusters for each team country than ideal, but some insight can be drawn. 
France has a small cluster in the 0-20 minute region, suggesting a number of players were barely used at all whilst still being named in match day squads.
Scotland have a fairly stronger cluster between 0 - 100 minutes, but then very little between 100-220. This suggests that player roles have been more consistent within the Scotland team as you either get a majority of the minutes or sub-100 minutes. 
Italy has a strong cluster in the 300-350 minute regions, suggesting a consistent core group of players.
Wales have a consistent distribution between 150-250 minutes, suggesting more turnover in the squad due to the limited nature of minutes played where these players are averaging between 2-3 full games.
England appear to show the least amounting of clustering along the distribution, combined with the fewest players used, suggests a fairly consistent squad was used throughout the Six Nations.
Ireland has the strongest cluster between 100-200 minutes, which suggests either consistency between substitute appearances or perhaps some experimentation of the squad players.

```python
top_15_metres_made = file_players[['Team','Player','Metres']].sort_values(by = 'Metres', ascending = False).head(15)
sns.barplot(data = top_15_metres_made, y = 'Player', x = 'Metres', hue = 'Team', dodge = False)
plt.title('Top 15 Carriers by Metres Carried')
plt.legend(bbox_to_anchor = (1,0.7))
plt.show()
```

The above visualisation shows the top 15 carriers for the entire Six Nations, colour coded by country. 
Ireland have the top 2 for most metres carried. Whilst England and Wales only have 1 each in the top 15, with Freddie Steward at 5 for England and Rio Dyer at 15 for Wales. 

```python
# Create a dataframe for most metres carried for each team
temp_dict = {}
for i in range(0, 6):
    temp = file_players.loc[(file_players['Metres'] == file_players.groupby('Team')['Metres'].max().values[i]) \
                       & (file_players['Team'] == file_players.groupby('Team')['Metres'].max().index[i])]

    temp_dict[i] = temp
#temp_dict
final_table_test = pd.concat(temp_dict, axis = 0, ignore_index=False)
final_table_test
```

```python
sns.barplot(data = final_table_test, x = 'Metres', y = 'Player', hue = 'Team', dodge = False)
plt.axvline(file_players['Metres'].mean(), linestyle = '--', alpha = 0.5, color = 'black', label = 'Mean (All)')
plt.axvline(final_table_test['Metres'].mean(), linestyle = '-', alpha = 0.5, color = 'black', label = 'Mean (Top)')
plt.legend(bbox_to_anchor = (1,0.5))
plt.title('Player with Most Metres Carried for Each Country')
plt.show()
```

The above visualisation show the top carrier for each country, with two separate mean values for comparison. The Mean (All) is the mean value for the entire population of players, whilst Mean (Top) only considers these 6 players. Italy and Wales players fall well below the average for these top players., but are clearly all well clear of the mean for the entire population.

```python
bottom_15_metres_made = file_players.query('`Mins Played` >= 80')[['Team','Player','Metres']].sort_values(by = 'Metres', ascending = False).tail(15)
sns.barplot(data = bottom_15_metres_made, y = 'Player', x = 'Metres', hue = 'Team', dodge = False)
plt.title('Bottom 15 Carriers by Metres Carried (80 mins min.)')
plt.legend(bbox_to_anchor = (1,0.5))
plt.show()
```

The above visualisation showcases the players with the least metres carried having played at least 80 mins. England and France only have one player each in this bottom 15, suggesting that ball carrying is non uniform for these countries. Conversely, Scotland have 6 players within this visualisation corresponding to the idea that Scotland carrying is more uniform across the population but is skewed due to the effectiveness of their dominant carriers.

```python
# Interested in metres per min after considering the bottom 15 carriers
# Created a dataframe of the best players for metres per min having played at least 80 mins
# Once again these players are mainly back 3 players apart from Marcus Smith who is a fly half
file_players['Metres per Min'] = file_players['Metres']/file_players['Mins Played']
temp_dict = {}
for i in range(0, 6):
    temp = file_players[file_players['Mins Played'] >= 80].loc[(file_players[file_players['Mins Played'] >= 80]['Metres per Min'] == file_players[file_players['Mins Played'] >= 80].groupby('Team')['Metres per Min'].max().values[i]) \
                       & (file_players[file_players['Mins Played'] >= 80]['Team'] == file_players[file_players['Mins Played'] >= 80].groupby('Team')['Metres per Min'].max().index[i])]

    temp_dict[i] = temp
top_metres_per_min = pd.concat(temp_dict, axis = 0, ignore_index=False)
top_metres_per_min
```

```python
sns.barplot(data = top_metres_per_min, x = 'Metres per Min', y = 'Player', hue = 'Team', dodge = False)
plt.axvline(file_players['Metres per Min'].mean(), linestyle = '--', alpha = 0.5, color = 'black', label = 'Mean (All)')
plt.axvline(top_metres_per_min['Metres per Min'].mean(), linestyle = '-', alpha = 0.5, color = 'black', label = 'Mean (Top)')
plt.title('Top Carrier per Min for each Country (80 mins min.)')
plt.legend(bbox_to_anchor = (1,0.5))
plt.show()
```

The above visualisation showcases the players with the best metres per min for each country having played at least 80 minutes.
Blair Kinghorn is the dominant player, with his metres per min exceeding the mean of the top 6 population.
Ange Capuozzo is the least effective of these six players, which suggests Italy shared ball carrying duties within the squad.

```python
top_15_metres_per_m = file_players.query('`Mins Played` >= 80')[['Team','Player','Metres per Min']].sort_values(by = 'Metres per Min', ascending = False).head(15)
sns.barplot(data = top_15_metres_per_m, y = 'Player', x = 'Metres per Min', hue = 'Team', dodge = False)
plt.title('Top 15 Carriers by Metres Carried per Min (80 mins min.)')
plt.legend(bbox_to_anchor = (1,0.5))
plt.show()
```

The above visualisation demonstrates the top 15 carriers for metres per min having played at least 80 minutes. France is the least well represented country within this group, suggesting Thomas Ramos probably dominated kick returns for France.
Edoardo Iachizzi of Italy is a flanker unlike all of the other players being backs, which makes his metres per min value very impressive as unlikely to be dominated by kick returns.

```python
bot_15_metres_per_m = file_players.query('`Mins Played` >= 80')[['Team','Player','Metres per Min']].sort_values(by = 'Metres per Min', ascending = False).tail(15)
sns.barplot(data = bot_15_metres_per_m, y = 'Player', x = 'Metres per Min', hue = 'Team', dodge = False)
plt.title('Bottom 15 Carriers by Metres Carried per Min (80 mins min.)')
plt.legend(bbox_to_anchor = (1,0.5))
plt.show()
```

The above visualisation demonstrates the bottom 15 carriers for metres per min having played at least 80 minutes. Scotland are once again well represented with 4 of the bottom 15 players, and this suggests that Grant Gilchrist's game does not focus on ball carrying to be useful.
The only two none forwards are both Ireland scrum-halfs, Craig Casey and Conor Murray. This will most likely be part tactical, as scrum-halfs primary job is to pass the ball away from a ruck which will not accumulate many carrying metres.
